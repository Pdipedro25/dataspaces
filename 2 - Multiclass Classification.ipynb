#### Function Definition

* *function*: plot_learning_curve()

given a set of parameters, such as estimator to be represented or train sizes for learning phases, it represent the evolution of learning stage in a two curves-based graph.

```python
def plot_learning_curve(estimator, title, X, y, ylim=None, cv=None,
                        n_jobs=None, train_sizes=np.linspace(.01, 1.0, 50)):
   
    plt.figure(figsize=(15,7))
    plt.title(title)
    if ylim is not None:
        plt.ylim(*ylim)
    plt.xlabel("Training examples")
    plt.ylabel("Score")
    train_sizes, train_scores, test_scores = learning_curve(
        estimator, X, y, cv=cv, n_jobs=n_jobs, train_sizes=train_sizes)
    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)
    plt.grid()

    plt.fill_between(train_sizes, train_scores_mean - train_scores_std,
                     train_scores_mean + train_scores_std, alpha=0.1,
                     color="r")
    plt.fill_between(train_sizes, test_scores_mean - test_scores_std,
                     test_scores_mean + test_scores_std, alpha=0.1, color="g")
    plt.plot(train_sizes, train_scores_mean, 'o-', color="r",
             label="Training score")
    plt.plot(train_sizes, test_scores_mean, 'o-', color="g",
             label="Cross-validation score")

    plt.legend(loc="best")
    return plt
```

* *function*: cross_validation()

function that made the GridSearch for the classifier (and with the setup parameters) and return the best estimator based on the Accuracy scores.

```python
def cross_validation(clf, clf_params,X,y):
    start = time.time()
    best_clf = GridSearchCV(estimator=clf,param_grid=clf_params, cv=5,scoring=None, return_train_score=True)
    best_clf.fit(X,y)
    end = time.time()
    print("Accuracy for result of cross validation:",best_clf.best_score_)
    print("Cross validation requires:",end-start,"sec")
    return best_clf.best_estimator_
```

* *function*: score_model()

helpful function that plot different statistics about the estimator such as time spent for both training and test and Accuracy and LogLoss score

```python
def score_model(clf,X,y,name):
    X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = TestSize,random_state=Seed1)
    start_tr = time.time()
    clf.fit(X_train,y_train)
    training_time=time.time()-start_tr
    start_te = time.time()
    predictions = clf.predict_proba(X)
    lloss=metrics.log_loss(y_true=y,y_pred=predictions,labels=[0,1,2,3,4,5,6])
    accuracy=clf.score(X_test,y_test)
    test_time = time.time()-start_te
    score_res = []
    score_res.append(("Name",name))
    score_res.append(("Accuracy",accuracy))
    score_res.append(("Log Loss",lloss))
    score_res.append(("Training Time",training_time))
    score_res.append(("Test Time",test_time))
    print("Scores:")
    for name,result in score_res:
        print(name,"=",result)
    return score_res
```

* *function*: pca_smart_comp()

made a calculation to retrieve the PCA components  

```python
def pca_smart_comp (pca_X,desired_perc):
    explained_variance = np.var(pca_X, axis=0)
    explained_variance_ratio = explained_variance / np.sum(explained_variance)
    cum=np.cumsum(explained_variance_ratio)
    print(cum)
    fi = plt.figure(figsize=(15,6))
    plt.plot(cum,'--o')
    plt.grid(True)
    plt.show()
    np.max(cum)
    desired=np.argmax(cum>desired_perc)
    print("Good Number of PC=",desired+1,"with Cumulative Variance =",cum[desired])
    return desired
```

* *function*: plot_multiclass_roc()

perform the Area Under ROC Curve calculation and visualization for multiclass estimator on the X (instance set) and y (label set) parameters.

```python
def plot_multiclass_roc(clf,X,y):

    lw = 1
    y = label_binarize(y, classes=[0,1,2,3,4,5,6])
    n_classes = y.shape[1]

    # shuffle and split training and test sets
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=TestSize,random_state=Seed1)

    # Learn to predict each class against the other
    classifier = OneVsRestClassifier(clf)
    y_score = classifier.fit(X_train, y_train).predict_proba(X_test)
    

    # Compute ROC curve and ROC area for each class
    fpr = dict()
    tpr = dict()
    roc_auc = dict()
    for i in range(n_classes):
        fpr[i], tpr[i], _ = roc_curve(y_test[:, i], y_score[:, i])
        roc_auc[i] = auc(fpr[i], tpr[i])

    # Compute micro-average ROC curve and ROC area
    fpr["micro"], tpr["micro"], _ = roc_curve(y_test.ravel(), y_score.ravel())
    roc_auc["micro"] = auc(fpr["micro"], tpr["micro"])
    
    if n_classes>2:

        # First aggregate all false positive rates
        all_fpr = np.unique(np.concatenate([fpr[i] for i in range(n_classes)]))

        # Then interpolate all ROC curves at this points
        mean_tpr = np.zeros_like(all_fpr)
        for i in range(n_classes):
            mean_tpr += interp(all_fpr, fpr[i], tpr[i])

        # Finally average it and compute AUC
        mean_tpr /= n_classes

        fpr["macro"] = all_fpr
        tpr["macro"] = mean_tpr
        roc_auc["macro"] = auc(fpr["macro"], tpr["macro"])

        # Plot all ROC curves
        plt.figure(figsize=(10,10))
        plt.plot(fpr["micro"], tpr["micro"],
                 label='micro-average ROC curve (area = {0:0.2f})'
                       ''.format(roc_auc["micro"]),
                 color='deeppink', linestyle=':', linewidth=4)

        plt.plot(fpr["macro"], tpr["macro"],
                 label='macro-average ROC curve (area = {0:0.2f})'
                       ''.format(roc_auc["macro"]),
                 color='navy', linestyle=':', linewidth=4)

        colors = cycle(['aqua', 'darkorange', 'cornflowerblue','green','red','yellow','magenta'])
        for i, color in zip(range(n_classes), colors):
            plt.plot(fpr[i], tpr[i], color=color, lw=lw,
                     label='ROC curve of class {0} (area = {1:0.2f})'
                     ''.format(i, roc_auc[i]))

        plt.plot([0, 1], [0, 1], 'k--', lw=lw)
        plt.xlim([0.0, 1.0])
        plt.ylim([0.0, 1.05])
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.title(clf)
        plt.legend(loc="lower right")
        plt.show()
```

## K-fold Cross Validation

__Cross Validation__ is the technique used to lead the model selection by overpassing the problem to be not able to estimate the test error rate: that's becouse, even if we have a Test Set, we cannot use it for model selection, becouse it will compromise his *purity* for model test and assestement phase. So in all the stages that precede the achievement of the model result, it is __avoid to use Test set__. 
For this reason a good way to approximate the Test set is to split another time the Training Set, to get a subset of instances, the _Validation Set_ used for Paramater Evaluation and/or for model selection. This is the main idea. __K-fold CV__ is a more complex application of this technique, that involves randomply dividing the set of observations into k groups, or folds, of approximately equal size: the first fold is treated as the Validation Set, and the model is fit on the remaining k-1 folds. The the __MSE__ (mean squared error) is used to measure how the model is working. This procedure is repeated k times; where each time the Validation Set is represented by a different folds, until everyone is used as ValSet. So, said the MSE:
$$MSE=\: \sigma \: ^{2}\: /\: n$$
The k-fold CV estimate is computed by averaging these values:
$$CV_{(k)}=\: \frac{1}{k}\sum_{1}^{k}\sigma \: ^{2}\: /\: n$$

The K-Fold CV is preferred to the Leave-One-Out Technique (that correspond to the K-Fold for k = # of instances) becouse of computational time: K-Fold is good compromise between the exhaustive research and a quick-computation research.

We will use CV with k=5 by usage of GridSearchCV method provided by SKLEARN library.

# Multiclass Classification

We will use as parameter of evaluation:

$Accuracy = \frac{TP + TN}{TP + FP + TN + FN}$
    
Accuracy, or Error Rate, describes the percentage of correct predictions made by the model. In short words, Accuracy = # correct / total predictions.


$Log Loss  =  -\sum_{c=1}^My_{o,c}\log(p_{o,c})$

where:
* N - number of observations
* M - number of possible class labels (dog, cat, fish)
* log - the natural logarithm
* y - a binary indicator (0 or 1) of whether class label c is the correct classification for observation o
* p - the model's predicted probability that observation o is of class c

we take the sum of log loss values for each class prediction in the observation.
Logarithmic loss measures the performance of a classification model where the prediction input is a probability value between 0 and 1. The goal of our machine learning models is to minimize this value. A perfect model would have a log loss of 0. _Log loss increases as the predicted probability diverges from the actual label._

$Training Time$

Time spent to _train_ the model

$Test Time$

Time spent to _use_ the model on testset


### Log Loss vs Accuracy

- ___Accuracy___ is the count of predictions where your predicted value equals the actual value. Accuracy is not always a good indicator because of its yes or no nature. Goal = get the _highest_ accuracy possible.
- ___Log Loss___ takes into account the uncertainty of your prediction based on how much it varies from the actual label. This gives us a deeper view into the performance of our model. Goal = get the _lowest_ logarithmic loss possible.

## Random Forest

__Random Forest__ is a supervised learning algorithm. It is an ensemble methods based on the generation of a number of Decision Trees (defined by an hyperparameter) on the training samples. Each of these are made by choosing, at each node of the tree, one feature from a subset of _m randomly selected features_, different at each step. It sounds illogical (we cannot choose the best one, but just one from the random subset) but it improves the predictive accuracy and controls *over-fitting,* particularly when there are strong predictors (so correlated to the prediction goal).

**Ensemble methods** are meta-algorithms that combine several machine learning techniques into one predictive model in order to decrease variance (bagging), bias (boosting), or improve predictions (stacking).

Random forest retrieve then the most reliable **Classification Tree** that can be extracted from the class object as *best estimator* parameter. A decision tree can be used to visually and explicitly represent decisions and decision making, representing a tree-like model of decisions.
The task of growing a classification tree is based on the use of *recursive
binary splitting* to grow a classification tree. Splitting phase needs a way to detect how a particular split is going to give improvements to the research or it is a trivial solution.

The splitting, so, is generally based on those *impurity measures*:

_classification error_ at a node t defined as
$Error(t)={1}-\max{j}{P}{\left({j}{|}{t}\right)}$
 
_gini index_ at a node t defined as
$\displaystyle{G}{I}{N}{I}{\left({t}\right)}={1}-∑{j}{P}{\left({j}{|}{t}\right)}{2}$
 
_entropy index_ at a node t defined as
$\displaystyle{E}{n}{t}{r}{o}{p}{y}{\left({t}\right)}=-∑{j}{P}{\left({j}{|}{t}\right)} \log{{\left({P}{\left({j}{|}{t}\right)}\right)}}$

where P(j|t) is the probability that a record at the node t belongs to the class j.

In particular, since classification error is not sufficiently sensitive for tree-growing, *Gini Index (or Entropy) is used in practice.* It is not hard to see that the Gini index takes on a small value if all of the ˆpmk’s are close to
zero or one. For this reason the Gini index is referred to as a measure of
node purity—a small value indicates that a node contains predominantly
observations from a single class.

A relevant issue about tree models is **overfitting**. This bad phenomenon may happen especially when we let the tree to become a fully-grown tree. **Random Forest** leads to perform, with the ensemble approach, a more generalized decision tree and generally prevent strong form of overfitting.

### Learning Curve

In a learning curve, the performance of a model both on the training and validation set is plotted as a function of the training set size.

**Learning curves** further allow to easily illustrate the concept of (statistical) bias and variance.

__Bias__ in this context refers to erroneous (e.g. simplifying) model assumptions, which can cause the model to _underfit_ the data, cause by approximation of a real-life problem. A high-bias model does not adequately capture the structure present in the data.

__Variance__ on the other hand quantifies how much the model varies as we change the training data. A high-variance model is very sensitive to small fluctuations in the training data, which can cause the model to overfit.

The amount of bias and variance can be estimated using __learning curves__: 
* A model exhibits __high variance, but low bias__ if the training score plateaus at a high level while the validation score at a low level, i.e., if there is a large gap between training and validation score.
* A model with __low variance but high bias__, in contrast, is a model where both training and validation score are low, but similar. Very simple models are high-bias, low-variance while with increasing model complexity they become low-bias, high-variance.

```python
plot_learning_curve(rf, "Learning Curve of the RandomForest", X_train, y_train, cv=5)
plt.show()
```

The _Learning Curve_ represent the trend of the training phase for each instance of Training. The red represent the _score_ (in this case, the Accuracy) on training set, the green the one reached in the cross-validation on the validation set.

In this case, it is clear that **Random Forest represent an  high variance-low bias model**, becouse of the large gap between curves, and it means that is very sensitive to small fluctuations in the training data, and problably it overfit. It is an obvious demonstration of the unflexibility of Decision Tree.

## PCA

Principal component analysis (PCA) is a statistical procedure that uses an orthogonal transformation to convert the variables of a dataset into a new set of variables which are linearly uncorrelated and called principal components. The principal components are ranked according to the variance of data along them. This technique can be used to reduce the dimensionality of the dataset by considering just the most important principal components.

So basically, PCA is a procedure that convert each instance of the dataset into its **principal components**: the first principal component has the largest possible variance, while  the succeeding ones in turn has the highest variance possible under the constraint that it is orthogonal to the preceding components. Variance, informally, measures how far a set of values are spread out from their average value.
So it, in short words, works as "summary" of the data designed to preserve the maximum informations possible with the dimensionality reduction choosen.

In order to reduce the dimensionality of a dataset $X$ with $n$ variables to a new dataset with $k$ variables using __PCA__ the following steps have to be followed:

* standardize the data
* calculate the Eigenvectors and Eigenvalues from the covariance matrix or correlation matrix, or perform Singular Vector Decomposition.
* sort the Eigenvectors according to their Eigenvalues in decreasing order
* build the $n×k$-dimensional projection matrix $W$ by putting the top $k$ Eigenvectors into the columns of $W$
* transform the dataset $X$ by multiplying it with $W$: $X_{t} = XW$
* We apply PCA to the dataset in order to reduce the number of features.

```python
npX = np.asarray(X)
npX = preprocessing.StandardScaler(copy=True,with_mean=True,with_std=True).fit_transform(npX) #Normalization
pca_X = decomposition.PCA(n_components=15,svd_solver ='full').fit_transform(npX)
print("PCA reduction shape:",pca_X.shape,"VS original shape:",X.shape)
```

The __total variance__ is the sum of variances of all individual principal components.
The fraction of __variance explained__ by a principal component is the ratio between the variance of that principal component and the total variance.
For several principal components, add up their variances and divide by the total variance.
So it means that the explained variance is a good measure of how the data are "well-resumed" by the PCA trasformation.


The most significant approach is to choose a reasonable value for cumulative variance and taking the principal components consequently; e.g. taking the first k eigenvectors that capture at least 90% of the total variance.

## K-Nearest Neighbor

Given a positive integer $K$ and a test observation $x$, the __KNN classifier__ first identifies the K points in the training data that are closest to $x$, represented by $N_{0}$. It then estimates the conditional probability for class $j$ as the fraction of points in $N_{0}$ whose response values equal $j$:

$$Pr(Y = j | X = x_{0})=\frac{1}{K}\sum_{i \epsilon N_{0}}I(y_{i} = j)$$

Finally, KNN applies Bayes rule and classifies the test observation to the class with the largest probability.
K-NN is a type of instance-based learning, or lazy learning, means that the model does not purely "learn" from training set, but measure some (k) distances from neighbours

__The KNN Algorithm:__
1. Load the data
2. Initialize K to your chosen number of neighbors
3. For each example in the data
    - Calculate the distance between the query example and the current example from the data.
    - Add the distance and the index of the example to an ordered collection
4. Sort the ordered collection of distances and indices from smallest to largest (in ascending order) by the distances

5. Pick the first K entries from the sorted collection

6. Get the labels of the selected K entries

7. If regression, return the mean of the K labels

8. If classification, return the mode of the K labels

Both for classification and regression, a useful technique can be used to assign weight to the contributions of the neighbors, so that the nearer neighbors contribute more to the average than the more distant ones.

In our CV, we will test several K values and two different approaches: weighted and standard.

_Parameters used for CrossValidation_:

* __n_neighbors__ : int, optional (default = 5) Number of neighbors to use by default for kneighbors queries.

* __weights__ : str or callable, optional (default = ‘uniform’) weight function used in prediction. Possible values: 
    - ‘uniform’ : uniform weights. All points in each neighborhood are weighted equally.
    - ‘distance’ : weight points by the inverse of their distance. in this case, closer neighbors of a query point will have a greater influence than neighbors which are further away.

## SVM

The __support vector machine__ is a generalization of a simple and intuitive classifier called the __maximal margin classifier__. Supposing that our data can be perfectly separated using a hyperplane, then there will in fact exist an infinite number of such hyperplanes. This is because a given separating hyperplane can usually be shifted a tiny bit up or down, or rotated, without coming into contact with any of the observations. The objective of the support vector machine algorithm is to find one hyperplane in an N-dimensional space(N — the number of features) that distinctly classifies the data points.

In order to construct a classifier based upon a separating hyperplane, we must have a reasonable way to decide which of the infinite possible separating hyperplanes to use. A natural choice is the __maximal margin hyperplane__ which is the separating hyperplane that is _farthest from the training observations_. The maximal margin hyperplane is the separating hyperplane for which the margin is largest.

__Support vectors__ are data points that are closer to the hyperplane and influence the position and orientation of the hyperplane. Using these support vectors, we maximize the margin of the classifier. Deleting the support vectors will change the position of the hyperplane. These are the points that help us build our SVM.

In addition to performing linear classification, SVMs can efficiently perform a non-linear classification using what is called the kernel trick, implicitly mapping their inputs into high-dimensional feature spaces. Kernel methods, or kernel trick, owe their name to the use of kernel functions, which enable them to operate in a high-dimensional, implicit feature space without ever computing the coordinates of the data in that space, but rather by simply computing the inner products between the images of all pairs of data in the feature space. This operation is often computationally cheaper than the explicit computation of the coordinates. It allows to transform the linear hyperplane to fit an un-linear problem.

We will use SVM in his linear version and using the RBF kernel function for an unlinear approach.

```python
linear_svm = svm.SVC(kernel="linear",probability=True)
best_linear_svm = cross_validation(linear_svm,clf_params=linear_svm_params,X=X_train,y=y_train)
```

### AUC ROC Curve

Let introduce AUC-ROC Curve, that here is just shown by way of example for SVM Classifiers, but in the third part will be used for evaluation. That's becouse it is a meaningfull evaluation technique but not so easy-to-understand for Multiclass classification problem. We will see it more deeply in Binary Classification Paragraph. 

__AUC (Area Under The Curve) ROC (Receiver Operating Characteristics) curve__ is one of the most important evaluation metrics for checking any classification model’s performance. To define this, we define the following two metrics:

True positive rate (TPR): this is the same as the __recall__: $TPR = \frac{TP}{FN + TP}$

False positive rate (FPR): this corresponds to the proportion of negative data points that are mistakenly considered as positive, with respect to all negative data points: $FPR = \frac{FP}{TN + FP}$

ROC is a probability curve and AUC represents degree or measure of separability. It tells how much model is capable of distinguishing between classes. Higher the AUC, better the model is at predicting the correct class for each istance.


The ROC curve is plotted with TPR against the FPR where TPR is on y-axis and FPR is on the x-axis.

An excellent model has AUC near to the 1 which means it has good measure of separability. A poor model has AUC near to the 0 which means it has worst measure of separability. In fact it means it is reciprocating the result.

_The AUC, briefly, represent the amount of probability that a single instance with state of nature "1" will be well classified (as "1") against the probability that a single instance with state of nature "0" will be wrongly classified as "1", so it is a king of measurement of capability of discerness between class-members vs other instances._
